```python
import pandas as pd
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import osmnx as ox
```

```python
# Importo usuasrios

# Load the CSV file into a DataFrame
df_registros = pd.read_csv('datasets/usuarios/post/usuarios_completo.csv')

# Convert 'fecha_alta' column to datetime type
df_registros['fecha_alta'] = pd.to_datetime(df_registros['fecha_alta'])

# Extract the year from 'fecha_alta' and create a new column 'year_alta' if not already present
if 'year_alta' not in df_registros.columns:
    df_registros['year_alta'] = df_registros['fecha_alta'].dt.year

# Calculate the quantity of users by year
users_by_year = df_registros['year_alta'].value_counts().reset_index()

# Rename the columns in the users_by_year DataFrame
users_by_year.columns = ['year', 'alta_usuarios']

# Sort the DataFrame by year
users_by_year = users_by_year.sort_values('year')

#Importo viajes por año

```

```python
import os
# Directory containing the CSV files
directory = 'datasets/recorridos/'

# Initialize an empty list to store the data
data = []

# Loop through the files in the directory
for filename in os.listdir(directory):
    if filename.endswith('.csv'):
        print(filename)
        # Extract the year from the filename
         # Extract the year from the filename
         # Extract the year from the filenam
        year = int(filename.split('-')[-1].split('.')[0])

        # Load the CSV file into a DataFrame
        df = pd.read_csv(os.path.join(directory, filename))

        # Calculate the length of the DataFrame as the amount of trips
        trips = len(df)

        # Append the year and amount of trips to the data list
        data.append({'year': year, 'amount_of_trips': trips})

# Create a DataFrame from the data list
trips_df = pd.DataFrame(data)

# Sort the DataFrame by year
trips_df = trips_df.sort_values('year')

```

```python
trips_df
```

```python
#Merge de dataframes

# Merge dataframes based on the 'Year' column
merged_df = pd.merge(users_by_year, trips_df, on='year', how='inner')
merged_df.to_csv('usuarios_viajes_ano.csv')
```
